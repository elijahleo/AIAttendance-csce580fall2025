```python
import pandas as pd
```

```python
df = pd.read_excel("AI_final_data.xlsx")
```

```python
# calculate median attendance
median_att= df['Attendance'].median()
print(f"Median attendance: {median_att}")
```

```python
# find class(es) with lowest attendance
min_att = df['Attendance'].min()
min_rows = df[df['Attendance'] == min_att]
print(f"Lowest attendance: {min_att}")
print(min_rows['Date'])
```

```python
# find class(es) with highest attendance
max_att = df['Attendance'].max()
max_rows = df[df['Attendance'] == max_att]
print(f"Highest attendance: {max_att}")
print(max_rows['Date'])
```

```python
# calculate mean attendance on assessment and non-assessment days
test_att = df.loc[df['Assessment'] == 1, 'Attendance'].mean()
non_test_att = df.loc[df['Assessment'] == 0, 'Attendance'].mean()
print(f"Attendance for assessments: {test_att}")
print(f"Attendance for non-assessment classes: {non_test_att}")
```

```python
# get attendance on each assessment day
print(df[df['Assessment'] == 1]['Attendance'])
```

```python
# calculate number of non-assessment classes with attendance higher than average for assessments
non_test_days = df[(df['Assessment'] == 0) & (df['Attendance'] >= test_att)]
print(len(non_test_days))
```

```python
# calculate number of non-assessment classes with attendance higher than average for assessment days in which whole class was assessed
# that is, excluding day of grad student presentations
non_test_days_exc = df[(df['Assessment'] == 0) & (df['Attendance'] >= 43)]
print(len(non_test_days_exc))
print(non_test_days_exc)
```
